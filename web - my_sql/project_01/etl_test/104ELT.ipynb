```python
import requests
import pandas as pd
from IPython.display import clear_output

# 與伺服器做連線
# 讓104 不要以為我們是爬蟲，於是新增 User-Agent 還有 Referer，Referer 指的是前一次造訪的網頁
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36",
    "Referer": "https://www.104.com.tw/"
}

url = "https://www.104.com.tw/jobs/search/list?ro=0&kwop=7&keyword=爬蟲工程師&expansionType=area,spec,com,job,wf,wktm&order=12&asc=0&page=4&mode=l&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1"

# 發送 GET 請求取得 JSON 格式的回應
resp = requests.get(url, headers=headers)

# 將取得的 JSON 格式資料轉成 DataFrame
df = pd.DataFrame(resp.json()["data"]['list'])

# 從 DataFrame 中選擇需要的欄位
selected_cols = ['jobName', 'jobNameSnippet', 'jobAddrNoDesc', 'jobAddress', 'custNameRaw', 'salaryDesc']
df = df[selected_cols]

# 將 DataFrame 寫入 CSV 檔案
df.to_csv('job_data.csv', index=False, encoding='utf-8-sig')

```
